```python
## Å analysere geojson-format blir nytt for meg!

geometryfile = 'geometri.geojson'
geodf = geopd.read_file(geometryfile).to_crs(epsg=25832) # Reprojected
```

```python
geodf_error = case.error_geometries(geodf, outfile=r'error_geometri.geojson', max_distance=1.0)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
geodf.geometry.plot(ax=ax, label='geometry')
geodf_error.geometry.plot(ax=ax,color = 'r', markersize=6, label='geometry-error')
plt.legend();plt.show()
```
